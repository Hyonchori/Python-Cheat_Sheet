# 1. 산술 연산자


| **연산자**   |      **의미**      |  **예** |
|----------|-------------|------|
| **+** |  양변의 값을 더하기 | a = 3 + 2 (a는 5) |
| **-** |  양변의 값을 빼기 | a = 3 - 2 (a는 1) |
|  *  |  양변의 값을 곱하기 | a = 3 * 2 (a는 6) |
| **/** |  좌변의 값을 우변의 값으로 나누기 | a = 3 / 2 (a는 1.5) |
| **//** |  좌변의 값을 우변의 값으로 나눈 몫 | a = 3 // 2 (a는 1) |
| **%** |  좌변의 값을 우변의 값으로 나눈 나머지 | a = 3 % 2 (a는 1) |
| **  |  좌변의 값을 우변의 값으로 제곱 | a = 3 ** 2 (a는 9) |

# 2. 대입 연산자
| **연산자**   |      **의미**      |  **예** |
|----------|-------------|------|
| **=** |  대입(할당) | a = 3|
| **+=** |  좌변의 변수에 우변의 값을 더한 값을 대입 | a += 2 (a는 5) |
| **-=** |  좌변의 변수에 우변의 값을 뺀 값을 대입 | a -= 2 (a는 1) |
| **x(별표)=** |  좌변의 변수에 우변의 값을 곱한 값을 대입 | a * = 2 (a는 6) |
| **/=** |  좌변의 값을 우변의 값으로 나눈 값을 대입 | a /= 2 (a는 1.5) |
| **//=** |  좌변의 값을 우변의 값으로 나눈 몫을 대입 | a //= 2 (a는 1) |
| **%=** |  좌변의 값을 우변의 값으로 나눈 나머지를 대입 | a %= 2 (a는 1) |
| **xx(별표2개)=** |  좌변의 값을 우변의 값으로 제곱한 값을 대입 | a ** = 2 (a는 9) |

# 3. 관계 연산자
| **연산자**   |      **의미**      |  **예** |
|----------|-------------|------|
| **==** |  양변의 값이 같으면 True 반환 | a, b = 3, 2 <br> a == b #False|  
| **!=** |  양변의 값이 다르면 True 반환 | a != b #True|
| **>** |  좌변의 값이 우변의 값보다 크면 True 반환 | a > b #True|
| **<** |  좌변의 값이 우변의 값보다 작으면 True 반환 | a < b #False|
| **>=** |  좌변의 값이 우변의 값보다 크거나 같으면 True 반환 | a >= b #True|
| **<=** |  좌변의 값이 우변의 값보다 작거나 같으면 True 반환 | a <= b #False|

---
# 4. 논리 연산자
| **연산자**   |      **의미**      |  **예** |
|----------|-------------|------|
| **and** |  양변의 값 모두 True일 때 True 반환 | a, b = True, False <br> a and b #False|  
| **or** |  양변의 값 중 하나라도 True일 때 True 반환 | a or b #True|
| **not** | True일 때 False 반환, False일 때 True 반환 | not b #True|

---
# 5. 비트 연산자


| **연산자**   |      **의미**      |  **예** |
|----------|-------------|------|
| **&** |  양변의 비트값 모두 1일 경우 1 반환 | a, b = 1, 0 <br> a & b #0|  
| ** ** |  양변의 값 중 하나라도 1일 때 1 반환 | a  b #1|
| **^** | 양변의 값이 다를 경우 1, 같을 경우 0을 반환 | a ^ b #1|
| **~** | 비트값이 1이면 0, 0이면 1을 반환 | ~b #1|
| **<<** | 좌변의 값을 우변의 값만큼 비트를 왼쪽으로 이동 |  x = 4 <br> a << 1 #8|
| **>>** | 좌변의 값을 우변의 값만큼 비트를 오른쪽으로 이동 | x = 4 <br> a >> 1  #2|

# 6. 연산자 우선 순위
- 괄호가 최우선
- 산술 연산이 비트 연산보다 우선
- 관계 연산이 논리 연산보다 우선
| **우선 순위**   |      **연산자**      |
|----------|-------------|
| **1** |  () |
| **2** |  ** |
| **3** |  +, -, ~ (부호)|
| **4** |  *, /, //, % |
| **5** |  +, - |
| **6** |  <<, >> |
| **7** |  & |
| **8** |  ^ |
| **9** |  or비트연산자 |
| **10** |  <, <=, >, >=, ==, != |
| **11** |  not |
| **12** |  and |
| **13** |  or |
